# Matplotlib example workflow

```python
# 0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x=[1,2,3,4]
y=[11,22,44,77]

# 2. Set up plot
fig, ax =plt.subplots(figsize=(10,10)) #(width, height)

# 3. plot the data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show fig
fig.savefig('../sample_plot.png')
```

#Which one should I use? (pyplot v matplotlib OO method?)

* When plotting something quickly use pyplot
* More advanced, OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
over_50.shape
```

```python
#pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# OO method
fig, ax =plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
ax.set_xlim([30,100]) #set limits on axes
```

```python
# OO method from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target']);

#Customize the plot
ax.set(title='Heart disease and cholesterol levels',
       xlabel='Age',
       ylabel='Cholesterol');
#add legend
ax.legend(*scatter.legend_elements(), title='Target');

#add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='-.')
```

```python
# OO method from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['thalach']);

#Customize the plot
ax.set(title='Heart disease and cholesterol levels',
       xlabel='Age',
       ylabel='Cholesterol');
#add legend
ax.legend(*scatter.legend_elements(), title='Target');

#add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='-.')
```

```python
over_50.head()
```

```python
# subplot of chol, age thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True);

#add data to ax0
scatter=ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target']);

#customize ax0
ax0.set(title='Heart disease and cholesterol levels',
        xlabel='Age',
        ylabel='Cholesterol');

#add legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target');

#add data to ax1
scatter=ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target']);
#customize ax1
ax1.set(title='Heart disease and max heart rate',
        xlabel='Age',
        ylabel='Max heart rate');

# add a legend
ax1.legend(*scatter.legend_elements(), title='Targe')
#add a meanline
ax1.axhline(y=over_50['thalach'].mean(), linestyle='-.');

#add title to figure
fig.suptitle('Heart disease Analysis', fontsize=16, fontweight='bold');
```

```python
#reset the style
plt.style.use('seaborn-whitegrid')
# OO method from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target'],
                   cmap='plasma'); #this changes the color scheme

#Customize the plot
ax.set(title='Heart disease and cholesterol levels',
       xlabel='Age',
       ylabel='Cholesterol');
#add legend
ax.legend(*scatter.legend_elements(), title='Target');

#add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='-.')
```

```python
# OO method from scratch
fig, ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['thalach'],
                   cmap='winter');

#Customize the plot
ax.set(title='Heart disease and cholesterol levels',
       xlabel='Age',
       ylabel='Cholesterol');
#add legend
ax.legend(*scatter.legend_elements(), title='Target');

#add a horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='-.')
```

```python
#customizing the x and y label limitations


# subplot of chol, age thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True);

#add data to ax0
scatter=ax0.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter');

#customize ax0
ax0.set(title='Heart disease and cholesterol levels',
        xlabel='Age',
        ylabel='Cholesterol');
ax0.set_xlim([50,80]) #change x-axis limits
#add legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target');

#add data to ax1
scatter=ax1.scatter(x=over_50['age'],
                    y=over_50['thalach'],
                    c=over_50['target'],
                    cmap='plasma');
#customize ax1
ax1.set(title='Heart disease and max heart rate',
        xlabel='Age',
        ylabel='Max heart rate');
ax1.set_xlim([50,80]) #change x-axis limits
ax1.set_ylim([60,200]) #change x-axis limits
# add a legend
ax1.legend(*scatter.legend_elements(), title='Targe')
#add a meanline
ax1.axhline(y=over_50['thalach'].mean(), linestyle='-.');

#add title to figure
fig.suptitle('Heart disease Analysis', fontsize=16, fontweight='bold');
```
